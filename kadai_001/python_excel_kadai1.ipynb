```python
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment, Border, Side
from datetime import datetime

# 現在日付取得
today_display = datetime.today().strftime("%Y/%m/%d")
today_file = datetime.today().strftime("%Y%m%d")

file_name = f"請求書_{today_file}.xlsx"


wb = Workbook()
ws = wb.active
ws.title = "請求書"


ws["A2"] = "請求書"
ws["A2"].font = Font(size=16, bold=True)
ws["A2"].alignment = Alignment(horizontal="center")


ws["A4"] = "株式会社ABC"
ws["A5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["A6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["A7"] = "担当者名: 鈴木一郎 様"


ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = today_display


ws["A10"] = "商品名"
ws["B10"] = "数量"
ws["C10"] = "単価"
ws["D10"] = "金額"


ws["A11"] = "商品A"
ws["B11"] = 2
ws["C11"] = 10000
ws["D11"] = "=B11*C11"

ws["A12"] = "商品B"
ws["B12"] = 1
ws["C12"] = 15000
ws["D12"] = "=B12*C12"


ws["D13"] = "=SUM(D11:D12)"

ws["A15"] = "合計"
ws["D15"] = "=D13"

ws["A16"] = "消費税"
ws["D16"] = "=D15*0.1"

ws["A17"] = "税込合計"
ws["D17"] = "=D15+D16"


thin = Side(style="thin")
border = Border(left=thin, right=thin, top=thin, bottom=thin)

for row in ws["A10:D12"]:
    for cell in row:
        cell.border = border


ws.column_dimensions["A"].width = 15
ws.column_dimensions["B"].width = 10
ws.column_dimensions["C"].width = 12
ws.column_dimensions["D"].width = 15
ws.column_dimensions["F"].width = 8
ws.column_dimensions["G"].width = 15


wb.save(file_name)

print(f"{file_name} を作成しました。")

```
